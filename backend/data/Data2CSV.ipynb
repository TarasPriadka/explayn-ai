```python
import pandas as pd
import os
```

```python
with open('161_links.txt') as f:
    links = [a.strip() for a in f.readlines()]
links
```

```python
texts = []
for i in range(len(links)):
    with open(f'txt/{i}.txt') as f:
        texts.append(f.read())
```

```python
df = pd.DataFrame({'link': links, 'text': texts})
df
```

```python
df.to_csv('161.csv', index=False)
```
